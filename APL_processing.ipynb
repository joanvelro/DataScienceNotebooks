```pyspark3
%%sql
SELECT TO_TIMESTAMP(Last_Update) AS last_update,
       Name,
       avg(Average_Speed) AS Average_Speed,
       sum(Count_) AS norm_volume,
       sum(Count_Short) AS Count_Short,
       sum(Count_Medium) AS Count_Medium,
       sum(Count_Long) AS Count_Long,
       avg(Average_Occupancy) AS Average_Occupancy,
       avg(Average_Headway) AS Average_Headway,
       avg(Average_length) AS Average_length,
       avg(Average_speed_short) AS Average_speed_short,  
       avg(Average_speed_medium) AS Average_speed_medium
FROM df_apl
GROUP BY Name, last_update 
```

```pyspark3
%%sql
DROP TABLE IF EXISTS APL_1MIN
```

```pyspark3
%%sql
CREATE TABLE APL_1MIN AS (
SELECT TO_TIMESTAMP(date_trunc('minute', Last_Update)) AS date_time_1min,
       Name,
       avg(Average_Speed) AS Average_Speed,
       sum(Count_) AS norm_volume,
       sum(Count_Short) AS Count_Short,
       sum(Count_Medium) AS Count_Medium,
       sum(Count_Long) AS Count_Long,
       avg(Average_Occupancy) AS Average_Occupancy,
       avg(Average_Headway) AS Average_Headway,
       avg(Average_length) AS Average_length,
       avg(Average_speed_short) AS Average_speed_short,  
       avg(Average_speed_medium) AS Average_speed_medium
FROM df_apl
GROUP BY date_time_1min
ORDER BY date_time_1min)
```
