## Bucles

```python
n = 0
while n < 10:
    print(n)
    n = n + 1
print('Fin')    
```

```python
x = 3
y = 0
z = x
while (z!=0): 
    y = y + x
    z = z -1
    print('x'+'='+str(x))
    print('y'+'='+str(y))
    print('z'+'='+str(z))
    print(str(x) + '+' + str(y) + '=' + str(y))
```

```python
x = 5
for i in range(0,x):
    print(i)
    print(str(i))
```

```python
x = 2
y = 3
for j in range(y):
    for i in range(x):
        print('x' + '=' + str(x))
        print("y" + "=" + str(y))
        print(str(i) + "=" + str(j))
```
