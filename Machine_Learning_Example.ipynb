# Logistic regression

```
from sklearn.datasets import load_iris
dataset = load_iris()
X, y = dataset.data, dataset.target

from sklearn import preprocessing
model = preprocessing.MinMaxScaler()
X = model.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```
#Import Library
from sklearn.linear_model import LogisticRegression
#Assumed you have, X (predictor) and Y (target)
#for training data set and x_test(predictor)
#of test_dataset
#Create logistic regression object
model = LogisticRegression()
#Train the model using the training sets
#and check score
model.fit(X_train, y_train)
model.score(X_train, y_train)
#Equation coefficient and Intercept
print('Coefficient: \n', model.coef_)
print('Intercept: \n', model.intercept_)
#Predict Output
y_hat =  model.predict(X_test)
```

# Naive Bayes 

```
from sklearn.datasets import load_iris
dataset = load_iris()
X, y = dataset.data, dataset.target

from sklearn import preprocessing
model = preprocessing.MinMaxScaler()
X = model.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```
#Import Library
from sklearn.naive_bayes import GaussianNB
#Assumed you have, X (predictor) and Y (target) for #training data set and x_test(predictor) of test_dataset #Create SVM classification object model = GaussianNB() #there is other distribution for multinomial classes like Bernoulli Naive Bayes
#Train the model using the training sets and check #score
model = GaussianNB()
model.fit(X_train, y_train)
print(model.sigma_)
#Predict Output
y_hat = model.predict(X_test)
```

```
plt.figure()
plt.plot(y_hat, color='r', label='prediction')
plt.plot(y_test, color='b', label='data test')
plt.legend()
plt.show()
```

```
plt.figure()
plt.scatter(y_test, y_hat)
plt.plot(np.arange(0,10),np.arange(0,10), '-k')
plt.show()
```

# kNN (K-Nearest Neighbours)

```
from sklearn.datasets import load_iris
dataset = load_iris()
X, y = dataset.data, dataset.target

from sklearn import preprocessing
model = preprocessing.MinMaxScaler()
X = model.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```
#Import Library
from sklearn.neighbors import KNeighborsClassifier #Assumed you have, X (predictor) and Y (target) for 
#training data set and x_test(predictor) of test_dataset 
#Create KNeighbors classifier object model KNeighborsClassifier(n_neighbors=6)
model = KNeighborsClassifier(n_neighbors=20, weights='uniform')
model.fit(X_train, y_train)
#default value for n_neighbors is 5
#Train the model using the training sets and check score model.fit(X, y)
#Predict Output
y_hat = model.predict(X_test)
```

```
plt.figure()
plt.plot(y_hat, color='r', label='prediction')
plt.plot(y_test, color='b', label='data test')
plt.legend()
plt.show()
```

```
plt.figure()
plt.scatter(y_test, y_hat)
plt.plot(np.arange(0,10),np.arange(0,10), '-k')
plt.show()
```
