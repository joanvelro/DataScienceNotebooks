```python
# Ejemplo de grafico de barras
myData1 = {'Genre':['male','female','male'],
           'City':[1,2,3],
           'Sold':[28,33,14]}
myData2 = {'Genre':['female','female','male'],
           'City':[1,2,3],
           'Sold':[8,53,24]}

df1 = pd.DataFrame(myData1, columns=['Genre', 'City', 'Age'])
df2 = pd.DataFrame(myData2, columns=['Genre', 'City', 'Age'])

df1['Key'] = 'trail1'
df2['Key'] = 'trail2'

DF = pd.concat([df1,df2],keys=['trail1','trail2'])

DFGroup = DF.groupby(['Genre','Key'])

DFGPlot = DFGroup.sum().unstack('Key').plot(kind='bar')
```
