## Data Preprocessing:

* Data Cleaning
* Data Smoothing
* Data Normalization/Standarization
* Data Aggrupation/Resample

```python
import pandas as pd
import glob
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt


#plt.style.use('fivethirtyeight')
plt.style.use('ggplot')
#plt.style.use('seaborn-colorblind')
#plt.style.use('seaborn-pastel')
plt.rcParams.update({'font.size': 12})
from matplotlib import rc
rc('font',**{'family':'serif','serif':['Times New Roman']})
rc('text', usetex=True)
```

```python
# %cd ..
# %ls
```

```python
%cd '/Volumes/HDD/data/ree/profiles'
```

```python
path = "/Volumes/HDD/data/ree/profiles/*.csv"
all_files = glob.glob(path)
cols = ['year', 'month', 'day', 'hour', 'season', 'coef_a', 'coef_b', 'coef_c',
       'coef_d']
df = pd.concat([pd.read_csv(f, sep=';', usecols=cols) for f in tqdm(all_files)])
```

```python
df.sort_values(by=['year', 'month', 'day', 'hour'], inplace=True)
print('{} info {}'.format('-'*20, '-'*20))
print(df.info())
print('{} head {}'.format('-'*20, '-'*20))
print(df.head())
print('{} tail {}'.format('-'*20, '-'*20))
print(df.tail())
df['year'] = df['year'].astype(int)
df['month'] = df['month'].astype(int)
df['day'] = df['day'].astype(int)
df['hour'] = df['hour'].map(lambda x: x -1)
```

```python
def check_nulls(df):
    """
    """
    if df.isna().sum().sum()>0:
        print('(!) Exist NaN')
        print(df.isna().sum())
        df.fillna(0, inplace=True)
    elif df.isnull().sum().sum()>0:
        print('(!) Exist NULLs')
        print(df.isnull().sum())
        df.fillna(0, inplace=True)
    else:
        print('Everything ok')
```

```python
check_nulls(df)
```

```python
def convert_to_date(row):
    """
    :param row:
    :return:
    """
    import datetime
    return datetime.datetime(int(row['year']), int(row['month']), int(row['day']))

def convert_to_timestamp(row):
    """
    :param row:
    :return:
    """
    import pandas as pd
    return pd.Timestamp(int(row['year']), int(row['month']), int(row['day']), int(row['hour']))
```

```python
df['timestamp'] = df.apply(lambda row: convert_to_timestamp(row), axis=1)
df['date'] = df.apply(lambda row: convert_to_date(row), axis=1)
```

```python
df['coef_a'] = 100000* df['coef_a']
df['coef_b'] = 100000* df['coef_b']
df['coef_c'] = 100000* df['coef_c']
df['coef_d'] = 100000* df['coef_d']
```

```python
df_aux = df[ (df['date']>'2018-04-15') & (df['date']<'2018-04-20') ]

fig, ax1 = plt.subplots(figsize=(14,4))
ax1.plot(df_aux['timestamp'].values, df_aux['coef_a'].values, '-')

fig, ax2 = plt.subplots(figsize=(14,4))
ax2.plot(df_aux['timestamp'].values, df_aux['coef_b'].values, '-')

fig, ax3 = plt.subplots(figsize=(14,4))
ax3.plot(df_aux['timestamp'].values, df_aux['coef_c'].values, '-')

fig, ax4 = plt.subplots(figsize=(14,4))
ax4.plot(df_aux['timestamp'].values, df_aux['coef_d'].values, '-')

plt.show()
```

## Smoothing

```python
window_length = 9
polyorder = 2

from scipy.signal import savgol_filter

df['coef_a_smooth'] = savgol_filter(df['coef_a'],window_length, polyorder )
df['coef_b_smooth'] = savgol_filter(df['coef_b'],window_length, polyorder )
df['coef_c_smooth'] = savgol_filter(df['coef_c'],window_length, polyorder )
df['coef_d_smooth'] = savgol_filter(df['coef_d'],window_length, polyorder )


df_aux = df[ (df['date']>'2018-04-15') & (df['date']<'2018-04-20') ]

fig, ax1 = plt.subplots(figsize=(14,4))
ax1.plot(df_aux['timestamp'].values, df_aux['coef_a'].values, 'o')
ax1.plot(df_aux['timestamp'].values, df_aux['coef_a_smooth'].values, '--')

fig, ax2 = plt.subplots(figsize=(14,4))
ax2.plot(df_aux['timestamp'].values, df_aux['coef_b'].values, 'o')
ax2.plot(df_aux['timestamp'].values, df_aux['coef_b_smooth'].values, '--')

fig, ax3 = plt.subplots(figsize=(14,4))
ax3.plot(df_aux['timestamp'].values, df_aux['coef_c'].values, 'o')
ax3.plot(df_aux['timestamp'].values, df_aux['coef_c_smooth'].values, '--')

fig, ax4 = plt.subplots(figsize=(14,4))
ax4.plot(df_aux['timestamp'].values, df_aux['coef_d'].values, 'o')
ax4.plot(df_aux['timestamp'].values, df_aux['coef_d_smooth'].values, '--')

plt.show()
```

## Normalization (0,1)

```python
for var in ['a','b','c','d']: 
    df['coef_{}_norm'.format(var)] = 0
    df['coef_{}_norm'.format(var)] = (df['coef_{}_smooth'.format(var)] - df['coef_{}_smooth'.format(var)].min()) / (df['coef_{}_smooth'.format(var)].max() - df['coef_{}_smooth'.format(var)].min())
    
df_aux = df[ (df['date']>'2018-04-15') & (df['date']<'2018-04-20') ]

fig, ax1 = plt.subplots(figsize=(14,4))
ax1.plot(df_aux['timestamp'].values, df_aux['coef_a_norm'].values, 'o--')
ax1.set_title('time series a) ')

fig, ax2 = plt.subplots(figsize=(14,4))
ax2.plot(df_aux['timestamp'].values, df_aux['coef_b_norm'].values, 'o--')
ax2.set_title('time series b) ')

fig, ax3 = plt.subplots(figsize=(14,4))
ax3.plot(df_aux['timestamp'].values, df_aux['coef_c_norm'].values, 'o--')
ax3.set_title('time series c) ')

fig, ax4 = plt.subplots(figsize=(14,4))
ax4.plot(df_aux['timestamp'].values, df_aux['coef_d_norm'].values, 'o--')
ax4.set_title('time series d) ')

plt.show()
```

## Standarization ( $\mu=0$; $\sigma=0$)

```python
for var in ['a','b','c','d']: 
    df['coef_{}_norm_z'.format(var)] = 0
    df['coef_{}_norm_z'.format(var)] = (df['coef_{}_smooth'.format(var)] - df['coef_{}_smooth'.format(var)].mean()) / (df['coef_{}_smooth'.format(var)].std())
    
df_aux = df[ (df['date']>'2018-04-15') & (df['date']<'2018-04-20') ]

fig, ax1 = plt.subplots(figsize=(14,4))
ax1.plot(df_aux['timestamp'].values, df_aux['coef_a_norm_z'].values, 'o--')

fig, ax2 = plt.subplots(figsize=(14,4))
ax2.plot(df_aux['timestamp'].values, df_aux['coef_b_norm_z'].values, 'o--')

fig, ax3 = plt.subplots(figsize=(14,4))
ax3.plot(df_aux['timestamp'].values, df_aux['coef_c_norm_z'].values, 'o--')

fig, ax4 = plt.subplots(figsize=(14,4))
ax4.plot(df_aux['timestamp'].values, df_aux['coef_d_norm_z'].values, 'o--')

plt.show()
```

## Aggrupate at daily level

```python
dfg = df.groupby(['year', 'month', 'day']).sum()[['coef_a_smooth', 'coef_b_smooth', 'coef_c_smooth', 'coef_d_smooth']].reset_index()
dfg['date'] = dfg.apply(lambda row: convert_to_date(row), axis=1)
print(dfg.head())

dfg_aux = dfg[ (dfg['date']>'2019-01-01') & (dfg['date']<'2020-01-01') ]

fig, ax1 = plt.subplots(figsize=(14,4))
ax1.plot(dfg_aux['date'].values, dfg_aux['coef_a_smooth'].values, '-')

fig, ax2 = plt.subplots(figsize=(14,4))
ax2.plot(dfg_aux['date'].values, dfg_aux['coef_b_smooth'].values, '-')

fig, ax3 = plt.subplots(figsize=(14,4))
ax3.plot(dfg_aux['date'].values, dfg_aux['coef_c_smooth'].values, '-')

fig, ax4 = plt.subplots(figsize=(14,4))
ax4.plot(dfg_aux['date'].values, dfg_aux['coef_d_smooth'].values, '-')
#ax1.set_title('Plot 2')
plt.show()
```

## Seasonal descomposition (hourly)

```python
df_aux = df[df['year']==2019]
index_time = pd.date_range(start=df_aux['timestamp'].min(),
                           end = df_aux['timestamp'].max(),
                           freq='H')
series = pd.DataFrame(data=df_aux['coef_a_smooth'].values,
                     index=index_time)

from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(series, model='additive')

#plt.figure(figsize=(20,4))
#result.plot()
#plt.show()


series_trend = result.trend
series_seasonal = result.seasonal
series_resid = result.resid
series_observed = result.observed

plt.figure(figsize=(14,4))
plt.plot(series_trend)
plt.show()

plt.figure(figsize=(14,4))
plt.plot(series_seasonal)
plt.show()

plt.figure(figsize=(14,4))
plt.plot(series_resid)
plt.show()

plt.figure(figsize=(14,4))
plt.plot(series_observed)
plt.show()
```

## Seasonal descomposition (daily)

```python
df_aux = dfg[dfg['year']>2016]

index_time = pd.date_range(start=df_aux['date'].min(),
                           end = df_aux['date'].max(),
                           freq='D')

series = pd.DataFrame(data=df_aux['coef_a_smooth'].values,
                     index=index_time)

from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(series, model='additive')

#plt.figure(figsize=(20,4))
#result.plot()
#plt.show()


series_trend = result.trend
series_seasonal = result.seasonal
series_resid = result.resid
series_observed = result.observed

plt.figure(figsize=(14,4))
plt.plot(series_trend)
plt.title('Trend')
plt.show()

plt.figure(figsize=(14,4))
plt.plot(series_seasonal)
plt.title('Seasonal')
plt.show()

plt.figure(figsize=(14,4))
plt.plot(series_resid)
plt.title('Redisual')
plt.show()

plt.figure(figsize=(14,4))
plt.plot(series_observed)
plt.title('Observed')
plt.show()
```
