```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# oferta
quantity = [10, 20, 30, 40, 50, 60, 80, 100]
precios_O = np.dot([50, 490, 1780, 4550, 7500, 12200, 19900, 35000] ,1/10000)
coefs_O = np.polyfit(quantity, precios_O, 2)
poly_O = np.poly1d(coefs_O)
quantity_p = np.linspace(0, 100, 30)
precios_Op = poly_O(quantity_p)

# demanda
precios_D = np.dot([35000, 19900, 12000, 7500, 4550, 1780, 490, 490],1/10000)
coefs_D = np.polyfit(quantity, precios_D, 3)
poly_D = np.poly1d(coefs_D)

precios_Dp = poly_D(quantity_p)

def demand(x, coefs):
  if len(coefs)==3:
    return coefs[0]*x**2 + coefs[1]*x + coefs[2]
  elif len(coefs)==4:
    return coefs[0]*x**3 + coefs[1]*x**2 + coefs[2]*x + coefs[3]
  else:
    return -1

def offer(x, coefs):
  if len(coefs)==3:
    return coefs[0]*x**2 + coefs[1]*x + coefs[2]
  elif len(coefs)==4:
    return coefs[0]*x**3 + coefs[1]*x**2 + coefs[2]*x + coefs[3]
  else:
    return -1

quantity_linear = np.linspace(0, 30, 30)
linear_demand = quantity_linear*(-0.12) + 5


#precio = costes**4
plt.figure(figsize=(12,6))
plt.scatter(quantity, precios_O, color='k', marker = 's', label='data offer')
plt.scatter(quantity, precios_D, color='k', marker='*', label='data demand')

plt.plot(quantity_p, precios_Op, color='b', label='fit offer')
plt.plot(quantity_p, precios_Dp, color='r', label='fit demand')

plt.plot(quantity_linear, linear_demand, color='c', label='demand (linear)')


plt.legend()
plt.xlabel('Quantity (X)')
plt.ylabel('Price (P)')
plt.grid()
plt.show()
```

```python
def cost(a0, x):
  return a0*np.log(x)

x = np.linspace(1, 100, 100)
costes1 = cost(10, x)
costes2 = cost(8, x)
costes3 = cost(12, x)


plt.figure(figsize=(12,6))
plt.plot(x, costes1, color='r', label='Costes $C = 10 \cdot \log(X) $')
plt.plot(x, costes2, color='b', label='Costes $C = 8 \cdot \log(X) $')
plt.plot(x, costes3, color='k', label='Costes $C = 12 \cdot \log(X) $')
plt.legend()
plt.xlabel('Quantity (X)')
plt.ylabel('Costs (C)')
plt.grid()
plt.show()
```
