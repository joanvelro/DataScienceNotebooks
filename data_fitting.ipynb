## Data fitting

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array((1.02, 0.95, 0.87, 0.77, 0.67, 0.56, 0.44, 0.30, 0.16, 0.01))
y = np.array((0.39, 0.32, 0.27, 0.22, 0.18, 0.15, 0.13, 0.12, 0.13, 0.15))
```

```python
y.shape
```

```python
X = np.concatenate((np.ones(len(x)), np.exp(x), np.exp(x)*x ))
X = np.concatenate((np.ones(len(x)), np.exp(x), np.exp(x)*x**2 ))
X = np.concatenate((np.ones(len(x)), np.exp(x), np.exp(x)*x**2, np.log(y), y**3))
X = np.concatenate((np.exp(y), y**3))
```

```python
X.shape
```

```python
Xm = np.matrix(X)
```

```python
beta = Xm.I*y
```

```python
beta.shape
```

```python
y_hat = Xm*beta
```

```python
y_hat.shape
```

```python
y_hat = np.array(y_hat.T)
```

```python
y_hat.shape
```

```python
y_hat = y_hat.ravel()
```

```python
resid = y-y_hat
```

```python
plt.boxplot(resid)
```

```python

plt.plot(x, y_hat, 'bo-')
plt.plot(x, y, 'r-o')
```
