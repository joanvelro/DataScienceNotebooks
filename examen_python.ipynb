## Pregunta 1.1
Crea una funcion a la cual le pases por parametros el 
numero de filas y el numero de columnas y te la rellene con numeros aleatorios del 1 al 100 con 2 decimales
Crea una matriz de 10 filas y 7 columnas. (No utilices NUMPY ni PANDAS en este apartado, en los demás puedes usarlo)

```
def create_random_matrix(nc, nr):
    """ create_random_matrix: creates a random matrix from 1 to 100 with two decimals
        :nc param: No. columns
        :nr param: no. rows
        :ouput: random matrix 
    """
    import random
    min_value = 1
    max_vlaue = 100
    return np.round(np.array([np.array(random.choices(range(min_value,max_vlaue), k=nc))*random.random() for _ in range(nr)]),2)
```

```
ncols = 7
nrows = 10
A = create_random_matrix(ncols, nrows)
print(A)
print(A.shape)
```

```
df = pd.DataFrame(data=A, columns=['Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes', 'Sabado', 'Domingo'])
print(df)  
```

## Pregunta 1.4
Normaliza todos los elementos de la primera matriz. Esto es, por
cada elemento resta la media y divide por la desviacion tipica del conjunto.

_Asumiendo que cada columna es una caracteristica:_

```
dfn = df.apply(lambda row: (row-np.mean(row))/np.std(row), axis=0)
print(dfn)
```

_Asumiento que cuando se refiere al conjunto, se refiere a toda la matriz:_

```
dfnp = df.apply(lambda row: (row-np.mean(df.values.tolist()))/np.std(df.values.tolist()), axis=0)
print(dfnp)
```

# BLOQUE 2: CREAR HISTOGRAMA Y DIAGRAMA DE BARRAS DE MÁS FRECUENTES.

## Pregunta 2.1
Crea un vector de números enteros aleatorios entre el 1 y el 10, de tamaño 100
Crea otro vector de numeros aleatorios entre el 11 y el 20 del mismo tamaño
y muestra por pantalla la tabla de frecuencia de la unión de los 2 vectores
para los valores 5,6 y 8 (de la primera) y para los valores 14,16,19 (de la segunda).

```
a = np.random.uniform(1,10,100).astype(int)
b = np.random.uniform(11,20,100).astype(int)
ab = np.concatenate([a,b])

print(ab)
```

```
table_a = pd.value_counts(a).reset_index().rename(columns={'index':'values',0:'Frequency'})
print(table_a[table_a['values'].isin([5,6,8])])
```

```
table_b = pd.value_counts(b).reset_index().rename(columns={'index':'values',0:'Frequency'})
print(table_b[table_b['values'].isin([14,16,19])])
```

# Pregunta 2.2
crea el histograma para dicho vector de frecuencias, con ancho de barra 0.75, título: histograma
y color de barra azul. Poner un título al eje x para diferenciar que son los números que se van a repetir.

```
plt.figure(figsize=(12,8))
plt.hist([a,b],rwidth=0.75, bins = np.arange(1,20,1))
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid()
plt.show()
```
