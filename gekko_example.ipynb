```python
from gekko import GEKKO
m = GEKKO() # Initialize gekko
m.options.SOLVER=1  # APOPT is an MINLP solver

# optional solver settings with APOPT
m.solver_options = ['minlp_maximum_iterations 500', \
                    # minlp iterations with integer solution
                    'minlp_max_iter_with_int_sol 10', \
                    # treat minlp as nlp
                    'minlp_as_nlp 0', \
                    # nlp sub-problem max iterations
                    'nlp_maximum_iterations 50', \
                    # 1 = depth first, 2 = breadth first
                    'minlp_branch_method 1', \
                    # maximum deviation from whole number
                    'minlp_integer_tol 0.05', \
                    # covergence tolerance
                    'minlp_gap_tol 0.01']
# Initialize variables
s1 = m.Var(value=50,lb=340,ub=5)
s2 = m.Var(value=45,lb=230,ub=5)
s3 = m.Var(value=35,lb=178,ub=5)
q1 = m.Var(value=1,lb=1,ub=500)
q2 = m.Var(value=1,lb=1,ub=500)
q3 = m.Var(value=1,lb=1,ub=500)
pq1 = 178
pq2 = 234
pq3 = 560
ps1 = 17
ps2 = 23
ps3 = 56
# Integer constraints for x3 and x4
b1 = m.Var(value=0,lb=0,ub=1,integer=True)
b2 = m.Var(value=0,lb=0,ub=1,integer=True)
b3 = m.Var(value=0,lb=0,ub=1,integer=True)
c1 = m.Var(value=0,lb=0,ub=1,integer=True)
c2 = m.Var(value=0,lb=0,ub=1,integer=True)
c3 = m.Var(value=0,lb=0,ub=1,integer=True)
m.Equation(q1**2 + q2**2 + q3**2<=500)
m.Equation(b1+b2+b3>=2)
m.Equation(c1+c2+c3>=2)
m.Obj(q1*pq1*b1 + q2*pq2*b2 +q2*pq2*b2+ s1*ps1*c1 + s2*ps2*c2 +s2*ps2*c3) # Objective
m.solve(disp=True) # Solve
print('q1: ' + str(q1.value))
print('q2: ' + str(q2.value))
print('q3: ' + str(q3.value))
print('s1: ' + str(s1.value))
print('s2: ' + str(s2.value))
print('s3: ' + str(s3.value))
print('Objective: ' + str(m.options.objfcnval))
```
