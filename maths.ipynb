```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,100000,0.1)

y1 = np.log(x)
y2 = np.log(y1)
y3 = np.log(y2)
y4 = np.log(y3)

plt.figure(figsize=(12,6))
plt.plot(x, y1, label='log(x)')
plt.plot(x, y2, label='log(log(x))')
plt.plot(x, y3, label='log(log(log(x)))')
plt.plot(x, y4, label='log(log(log(log(x))))')

plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.arange(0,10,0.1)


y1 = np.exp(0.5*x)
y2 = np.exp(0.4*x)
y3 = np.exp(0.2*x)
y4 = np.exp(0.6*x)

plt.figure(figsize=(12,6))
plt.plot(x, y1, label='$e^{a\cdot x}$ $a=0.5$')
plt.plot(x, y2, label='$e^{a\cdot x}$ $a=0.4$')
plt.plot(x, y3, label='$e^{a\cdot x}$ $a=0.2$')
plt.plot(x, y4, label='$e^{a\cdot x}$ $a=0.6$')
plt.grid(True)
plt.legend()
plt.show()

```
