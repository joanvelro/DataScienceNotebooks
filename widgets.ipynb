```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

def f(m, b, c):
    plt.figure(2)
    x = np.linspace(-10, 10, num=1000)
    y = m*x**2 +b*x + c + np.exp(x) 
    plt.plot(x, y)
    plt.scatter(np.random.normal(0,2,10), np.random.normal(0,2,10))
    plt.grid(True)
    plt.ylim(-5, 5)
    plt.show()

interactive_plot = interactive(f, m=(-3.0, 0), b=(-3, 3, 0.5), c = (-5, 5, 0.5))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot

```

```python
def f(label,m, b, c):
    plt.figure(2)
    x = np.linspace(-10, 10, num=1000)
    if label=='one':
      y = m*x**2 +b*x + c + np.exp(x) 
    elif label=='two':
      y = m*x**2 +b*x 
    plt.plot(x, y, label=label)
    plt.legend()
    plt.scatter(np.random.normal(0,2,10), np.random.normal(0,2,10))
    plt.grid(True)
    plt.ylim(-5, 5)
    plt.show()

interactive_plot = interactive(f, label=['one','two'], m=(-3.0, 0), b=(-3, 3, 0.5), c = (-5, 5, 0.5))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```
